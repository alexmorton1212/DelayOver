```python
import pandas as pd
import numpy as np
import os

df = pd.read_csv('../data/processed/xgb_model_cv_eval_results_with_thresholds.csv')

df
```

```python
score_cols = ['model_label', 'threshold', 'learning_rate', 'scale_pos_weight', 'subsample', 'recall_1', 'recall_0', 'precision_1', 'precision_0', 'roc_auc', 'percent_predicted_not_delayed']

df_boundary = df[score_cols].copy()
df_boundary['percent_not_delayed'] = df_boundary['percent_predicted_not_delayed'] / 100
df_boundary.drop('percent_predicted_not_delayed', axis=1, inplace=True)

df_boundary['score'] = (0.55 * df_boundary['recall_1']) + (0.05 * df_boundary['precision_0']) + (0.35 * df_boundary['recall_0']) + (0.05 * df_boundary['roc_auc'])
```

```python

boundary_model = df_boundary.sort_values(by='score', ascending=False)
```
