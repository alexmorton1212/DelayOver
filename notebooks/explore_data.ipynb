```python
import pandas as pd
import os

# Path to your parquet files
data_dir = "../data/raw"

# List all parquet files
files = [f for f in os.listdir(data_dir) if f.endswith('.parquet')]
print("Available parquet files:", files)

# Load all three parquet files into a single DataFrame
dfs = [pd.read_parquet(os.path.join(data_dir, f)) for f in files]
df_all = pd.concat(dfs, ignore_index=True)

df_all = df_all.dropna(axis=1, how='all')

# Quick peek at data
print(df_all.shape)
df_all.head()
```

```python
df_remove_cols = df_all[['Month', 'DayofMonth', 'DayOfWeek', 'OriginAirportID', 'DestAirportID', 'Reporting_Airline', 'OriginState', 'DestState', 'Flight_Number_Reporting_Airline', 'CRSDepTime', 'CRSArrTime', 'ArrTime', 
    'CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay', 'ArrDelayMinutes']]

df_remove_dups = df_remove_cols.copy().drop_duplicates()

df_remove_dups
```

```python
df_remove_dups[df_remove_dups['CarrierDelay'].notna()]
```

```python
df_0s = df_remove_dups.copy()

columns_to_fill = ['CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']
df_0s[columns_to_fill] = df_0s[columns_to_fill].copy().fillna(0)

df_0s[df_0s['OriginState'].notna()]
```

```python
state_codes = ['AK','AL','AR','AZ','CA','CO','CT','DC','DE','FL','GA','HI','IA','ID','IL','IN',
    'KS','KY','LA','MA','MD','ME','MI','MN','MO','MS','MT','NC','ND','NE','NH','NJ','NM','NV','NY',
    'OH','OK','OR','PA','RI','SC','SD','TN','TX','UT','VA','VT','WA','WI','WV','WY']

df_state_filter = df_0s[df_0s['OriginState'].isin(state_codes) & df_0s['DestState'].isin(state_codes)].copy()

df_state_filter
```

```python
df_state_filter['DestAirportID'].value_counts().head(200)
```

```python
import numpy as np
```

```python
df_state_filter.loc[:, 'delay_bin'] = np.where(df_state_filter['ArrDelayMinutes'] == 0, 'N', 'Y')
```

```python

# Count how many times each airport appears as origin
origin_counts = df_state_filter['OriginAirportID'].value_counts()

# Count how many times each airport appears as destination
dest_counts = df_state_filter['DestAirportID'].value_counts()

# Combine counts by adding them together, fill missing with 0 before adding
combined_counts = origin_counts.add(dest_counts, fill_value=0)

# Get top 200 airports by combined count
top_200_airports = combined_counts.nlargest(200).index

# Filter rows where both origin and destination are in the top 200 airports
airport_df = df_state_filter[
    (df_state_filter['OriginAirportID'].isin(top_200_airports)) &
    (df_state_filter['DestAirportID'].isin(top_200_airports))
]

airport_df
```

```python
summary = airport_df.groupby(['OriginAirportID', 'DestAirportID', 'Reporting_Airline', 'Month', 'DayOfWeek', 'delay_bin']).agg(
    delay_sum=('ArrDelayMinutes', 'sum'),
    count=('ArrDelayMinutes', 'count')
).reset_index()

summary

summary.to_csv('summary.csv', index=False)
summary.to_parquet('summary.parquet', index=False)
```
